```python
import os
import numpy as np
import matplotlib.pyplot as plt
import torchvision.utils as vutils
import torch
import PIL

def get_png_files(folder_path):
    png_files = []
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.png'):
            png_files.append(os.path.join(folder_path, file_name))
    return png_files

def read_image(image_path, target_size):
    img = PIL.Image.open(image_path)
    gray_img = img.convert("L")
    resized_img = gray_img.resize(target_size)
    img_array = np.array(resized_img)
    img_tensor = torch.from_numpy(img_array).unsqueeze(0)
    return img_tensor

def visualize(images):
    img_grid = vutils.make_grid(images, nrow=5)
    # Convert the image grid to a numpy array for plotting
    img_grid_np = np.transpose(img_grid.cpu().numpy(), (1, 2, 0))
    # Display the image grid using matplotlib
    plt.figure(figsize=(20, 20))
    plt.imshow(img_grid_np, cmap='gray')
    plt.axis("off")
    plt.show()

# Example usage
folder_path = "E:/kaggle/RSNAScreeningMammography/data/selection_transformed_data_original"
png_files = get_png_files(folder_path)
target_size = (347, 260)
images = [read_image(png_file, target_size) for png_file in png_files]
images = torch.stack(images, dim=0)
visualize(images)

```
